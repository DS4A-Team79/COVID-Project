```python
import numpy as np # linear algebra
import pandas as pd # data processing, I/O
```

```python
data = pd.read_csv("../Datasets/Raw/C19_CASES_DEATHS/time_series_covid_19_confirmed_US.csv")
byState = data.groupby("Province_State").sum()
print(byState[:5])
```

```python
data_cols = list(byState.columns)
print(data_cols[:20])
```

```python
byState.iloc[:,[0]]
```

```python
#get all dates at the end of a month between Jan 2020 and May 2021 in a list
month_dates = ['1/31/20','2/29/20','3/31/20','4/30/20',
'5/31/20','6/30/20','7/31/20','8/31/20',
'9/30/20','10/31/20','11/30/20','12/31/20',
'1/31/21','2/28/21','3/31/21','4/30/21',
'5/2/21']
```

```python
print(data_cols.index(month_dates[0]))
```

```python
end_month_cols = []

for month in month_dates:
    end_month_cols.append(data_cols.index(month))

print(end_month_cols)
```

```python
#Limit data to end dates
byState_Month = byState.iloc[:,end_month_cols]
byState_Month.head()
```

```python
#Prepare to remove non-states from data
non_states = ["American Samoa", "Guam", "District of Columbia", "Grand Princess", "Puerto Rico", "Virgin Islands", "Northern Mariana Islands", "Diamond Princess"]
data_rows = list(byState_Month.index)
print(data_rows[:5])
```

```python
not_states = []
for ns in non_states:
    not_states.append(data_rows.index(ns))
print(not_states)

```

```python
#remove non-states from data, and organize (transpose) it for visualizations
onlyState_month = byState_Month.copy()
onlyState_month = onlyState_month[~onlyState_month.index.isin(non_states)].T.reset_index()
onlyState_month.head()
```

```python
#Clean the data for exporting
cases_TS = onlyState_month.rename(columns={'index':'Date'})
cases_TS['Date'] = pd.to_datetime(cases_TS['Date'])
cases_TS = cases_TS.set_index("Date")
cases_TS["Month"] = cases_TS.index.to_period("M")
cases_TS_Month = cases_TS.pop("Month")
cases_TS.insert(0,"Month",cases_TS_Month)
cases_TS = cases_TS.reset_index().drop("Date",1)
cases_TS.head()
```

```python
cases_TS.to_csv("../Datasets/Refined_Datasets/state_cases_TS.csv")
```

```python
cols = list(cases_TS.columns)
cols = cols[1:]
print(cols[:5])
```

```python
state_cases_TS = pd.melt(cases_TS,id_vars=['Month'],value_vars=cols)
state_cases_TS = state_cases_TS.rename(columns={'Province_State':'State','value':'Total Cases'})
state_cases_TS.head()
```

```python
state_cases_TS.to_csv("../Datasets/Refined_Datasets/state_cases_monthly.csv")
```
