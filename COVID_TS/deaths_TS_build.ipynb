```python
import numpy as np # linear algebra
import pandas as pd # data processing, I/O
```

```python
data = pd.read_csv("../Datasets/Raw/C19_CASES_DEATHS/time_series_covid_19_deaths_US.csv")
byState = data.groupby("Province_State").sum()
print(byState[:5])
```

```python
data_cols = list(byState.columns)
print(data_cols[:20])
```

```python
byState.iloc[:,[0]]
```

```python
#get all dates at the end of a month between Jan 2020 and May 2021 in a list
month_dates = ['1/31/20','2/29/20','3/31/20','4/30/20',
'5/31/20','6/30/20','7/31/20','8/31/20',
'9/30/20','10/31/20','11/30/20','12/31/20',
'1/31/21','2/28/21','3/31/21','4/30/21',
'5/2/21']
```

```python
print(data_cols.index(month_dates[0]))
```

```python
end_month_cols = []

for month in month_dates:
    end_month_cols.append(data_cols.index(month))

print(end_month_cols)
```

```python
#Limit data to end dates
byState_Month = byState.iloc[:,end_month_cols]
byState_Month.head()
```

```python
#Prepare to remove non-states from data
non_states = ["American Samoa", "Guam", "District of Columbia", "Grand Princess", "Puerto Rico", "Virgin Islands", "Northern Mariana Islands", "Diamond Princess"]
data_rows = list(byState_Month.index)
print(data_rows[:5])
```

```python
not_states = []
for ns in non_states:
    not_states.append(data_rows.index(ns))
print(not_states)

```

```python
#remove non-states from data, and organize (transpose) it for visualizations
onlyState_month = byState_Month.copy()
onlyState_month = onlyState_month[~onlyState_month.index.isin(non_states)].T.reset_index()
onlyState_month.head()
```

```python
#Clean the deaths data for exporting
deaths_TS = onlyState_month.rename(columns={'index':'Date'})
deaths_TS['Date'] = pd.to_datetime(deaths_TS['Date'])
deaths_TS = deaths_TS.set_index("Date")
deaths_TS["Month"] = deaths_TS.index.to_period("M")
deaths_TS_Month = deaths_TS.pop("Month")
deaths_TS.insert(0,"Month",deaths_TS_Month)
deaths_TS = deaths_TS.reset_index().drop("Date",1)
deaths_TS.head()
```

```python
deaths_TS.to_csv("../Datasets/Refined_Datasets/state_deaths_TS.csv")
```

```python
cols = list(deaths_TS.columns)
cols = cols[1:]
```

```python
state_deaths_TS = pd.melt(deaths_TS,id_vars=['Month'],value_vars=cols)
state_deaths_TS = state_deaths_TS.rename(columns={'Province_State':'State','value':'Total Deaths'})
state_deaths_TS.head()
```

```python
state_deaths_TS.to_csv("../Datasets/Refined_Datasets/state_deaths_monthly.csv")
```

```python
state_new_deaths_TS = state_deaths_TS.copy()
state_new_deaths_TS.sort_values(by=['State','Month'],inplace=True)

#shift case numbers down one
state_new_deaths_TS['Old State'] = state_new_deaths_TS['State'].shift(1)
state_new_deaths_TS['Old Deaths'] = state_new_deaths_TS['Total Deaths'].shift(1)

#exclude data where current and old state don't match, and remove empty cell rows
state_new_deaths_TS = state_new_deaths_TS[state_new_deaths_TS['State'] == state_new_deaths_TS['Old State']]
state_new_deaths_TS.dropna(inplace=True)

#calculate total new deaths
state_new_deaths_TS['New Deaths'] = state_new_deaths_TS['Total Deaths'] - state_new_deaths_TS['Old Deaths']

#drop unnecessary columns to prep for exporting
state_new_deaths_TS.drop(columns=['Total Deaths','Old State','Old Deaths'],inplace=True)
state_new_deaths_TS.head()
```

```python
state_new_deaths_TS.to_csv("../Datasets/Refined_Datasets/state_new_deaths_monthly.csv")
```
